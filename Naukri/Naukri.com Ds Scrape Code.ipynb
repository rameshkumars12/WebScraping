```python
#Importing Libraries
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd


url = "https://www.naukri.com/data-science-jobs-in-hyderabad-secunderabad?k=data%20science&l=hyderabad%2Fsecunderabad"

#Getting the page links
Page_Links = []
for index,link in enumerate(url):
    if index == 9:
        break
    else:
        driver = webdriver.Chrome("D:\\ChromeD\\chromedriver.exe")
        driver.get(url)
        time.sleep(3)
        soup = BeautifulSoup(driver.page_source,"html5lib")
        driver.close()
        
        page = soup.find(class_ = "pagination mt-64 mb-60")
        next_page = page.find(class_ = "fright fs14 btn-secondary br2")["href"]
        base_path = f"https://www.naukri.com{next_page}"
        url = base_path
        Page_Links.append(url)
```

```python
#Getting the post details from range of pages
Get_Details = []
for urls in Page_Links:
    driver = webdriver.Chrome("D:\\ChromeD\\chromedriver.exe")
    driver.get(urls)
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source,"html5lib")
    driver.close()
    get_list = soup.find(class_ = "list")
    
    for i in get_list.find_all("article", class_ = "jobTuple bgWhite br4 mb-8"):
        #getting the job position
        position = i.find(class_ = "title fw500 ellipsis").get_text()     
        
        #getting the company name
        company = i.find(class_ = "subTitle ellipsis fleft").get_text()
        
        #getting the company rating
        try:
            rating = i.find(class_ = "starRating fleft dot").get_text()
        except:
            rating = "Nan"
        
        #getting the experience needed
        try:
            experience = i.find(class_ = "subTitle ellipsis fleft").get_text()
        except:
            experience = "Nan"
        
        #getting the salary details
        try:
            salary = i.find(class_ = "fleft grey-text br2 placeHolderLi salary").get_text()
        except:
            salary = "Nan"
        
        try:
            location = i.find(class_ = "fleft grey-text br2 placeHolderLi location").get_text()
            
        except:
            location = "Nan"
        
        #getting skills details
        try:
            skills = i.find(class_ = "tags has-description").get_text()
        except:
            skills = "Nan"
        
        
        #getting the post link
        try:
            link = i.find(class_ = "title fw500 ellipsis").get("href")
        except:
            link = "NaN"
            

        #Storing the data in a dictionary
        Details = {"Company": company,
                    "Rating" : rating,
                    "Position" : position,
                    "Experience" : experience,
                    "Salary" : salary,
                    "Location" : location,
                    "Skills" : skills,
                    "Link" : link
                }
        #Appending the Dictionary to the List
        Get_Details.append(Details)
```

```python
#Saving the list to DataFrame
df = pd.DataFrame(Get_Details)
```

```python
#Creating a CSV file
df.to_csv("Hyderabad DS Job Dataset.csv")
df.head(30)
```
