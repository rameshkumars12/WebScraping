```python
#Importing Libraries
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd


Get_Details = []
for x in range(2,5):
    
    link = f'https://www.naukri.com/data-science-jobs-in-hyderabad-{x}?k=data%20science&l=hyderabad'
    driver = webdriver.Chrome("D:\\ChromeD\\chromedriver.exe")
    driver.get(link)
    time.sleep(50)
    soup = BeautifulSoup(driver.page_source,"html5lib")
    #getting the list of contents
    driver.close()
    
    get_list = soup.find(class_ = "list")
    
    for i in get_list.find_all("article", class_ = "jobTuple bgWhite br4 mb-8"):
        
        #getting the job position
        position = i.find(class_ = "title fw500 ellipsis").get_text()     
        
        #getting the company name
        company = i.find(class_ = "subTitle ellipsis fleft").get_text()
        
        #getting the company rating
        try:
            rating = i.find(class_ = "starRating fleft dot").get_text()
        except:
            rating = "Nan"
        
        #getting the experience needed
        try:
            experience = i.find(class_ = "subTitle ellipsis fleft").get_text()
        except:
            experience = "Nan"
        
        #getting the salary details
        try:
            salary = i.find(class_ = "fleft grey-text br2 placeHolderLi salary").get_text()
        except:
            salary = "Nan"
        
        try:
            location = i.find(class_ = "fleft grey-text br2 placeHolderLi location").get_text()
            
        except:
            location = "Nan"
        
        #getting skills details
        try:
            skills = i.find(class_ = "tags has-description").get_text()
        except:
            skills = "Nan"
        
        
        #getting the post link
        try:
            link = i.find(class_ = "title fw500 ellipsis").get("href")
        except:
            link = "NaN"
            

        
        Details = {"Company": company,
                    "Rating" : rating,
                    "Position" : position,
                    "Experience" : experience,
                    "Salary" : salary,
                    "Location" : location,
                    "Skills" : skills,
                    "Link" : link
                }
        Get_Details.append(Details)

        print(Get_Details)
```
