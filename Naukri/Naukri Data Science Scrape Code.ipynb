### Step 1: Importing Libraries

```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
import pandas as pd


url = "https://www.naukri.com/data-science-jobs-in-bangalore-bengaluru?k=data%20science&l=bangalore%2Fbengaluru"
```

### Step 2: Getting Page Links

```python
#Getting the page links
Page_Links = []
for index,link in enumerate(url):
    if index == 20:
        break
    else:
        driver = webdriver.Chrome("D:\\ChromeD\\chromedriver.exe")
        driver.get(url)
        time.sleep(10)
        soup = bs(driver.page_source,"html5lib")
        driver.close()
        
        page = soup.find(class_ = "pagination mt-64 mb-60")
        next_page = page.find(class_ = "fright fs14 btn-secondary br2")["href"]
        base_path = f"https://www.naukri.com{next_page}"
        url = base_path
        Page_Links.append(url)
```

### Step 3: Scraping the Webpage

```python
#Getting the post details from range of pages
Get_Details = []
for urls in Page_Links:
    driver = webdriver.Chrome("D:\\ChromeD\\chromedriver.exe")
    driver.get(urls)
    time.sleep(3)
    soup = bs(driver.page_source,"html5lib")
    driver.close()
    get_list = soup.find(class_ = "list")


    for get_post in get_list.find_all(class_ = "jobTuple bgWhite br4 mb-8"):

        #getting the title
        try:
            get_position = get_post.find(class_ = "title fw500 ellipsis").get_text()
        except:
            get_position = " "

        #getting the company name
        try:
            get_company = get_post.find('a',class_='subTitle ellipsis fleft').get_text()
        except:
            get_company = " "

        #getting company rating
        try:
            get_rating = get_post.find('span',class_='starRating fleft dot').get_text()
        except:
            get_rating = " "
        
        #getting company review count
        try:
            get_review = get_post.find('a',class_='reviewsCount ml-5 fleft blue-text').get_text()
        except:
            get_review = " " 
          
        #getting the experience needed
        try:
            experience = get_post.find('li',class_='fleft grey-text br2 placeHolderLi experience')
            get_experience = experience.find('span',class_='ellipsis fleft fs12 lh16').get_text()
        except:
            get_experience = " "

        #getting the salary details 
        try:
            get_salary = get_post.find(class_ = "fleft grey-text br2 placeHolderLi salary").get_text()
        except:
            get_salary = " "

        #getting the location details
        try:
            get_location = get_post.find(class_ = "fleft grey-text br2 placeHolderLi location").get_text()
        except:
            get_location = " "

        #getting skills required
        try:
            get_skills = get_post.find(class_ = "tags has-description" ).get_text(",", strip = True)
        except:
            get_skills = " "

        #getting the post link
        try:
            get_link = get_post.find(class_ = "title fw500 ellipsis").get("href")
        except:
            link = " "



        #Storing the data in a dictionary
        Details = {"Company": get_company,
                    "Rating" : get_rating,
                   "Review Count" : get_review, 
                    "Position" : get_position,
                    "Experience" : get_experience,
                    "Salary" : get_salary,
                    "Location" : get_location,
                    "Skills" : get_skills,
                    "Link" : get_link
                    }    
        Get_Details.append(Details)
```

### Step 4: Converting List to DataFrame using Pandas

```python
df = pd.DataFrame(Get_Details)
df.head(5)
```
